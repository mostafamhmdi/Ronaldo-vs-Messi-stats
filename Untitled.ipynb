```python
import pandas as pd
```

```python
df=pd.read_excel('ronaldo_match_history.xlsx')
df1=pd.read_excel('messi_match_history.xlsx')
```

```python
df['Name']='Ronaldo'
df1['Name']='Messi'
```

```python
df1
```

```python
pre1 = 'r_'
pre2 = 'm_'

df['match_no']=df['match_no'].astype('str')
df1['match_no']=df1['match_no'].astype('str')
# Adding prefix 'r_' to 'match_no' column in 'df'
df['match_no'] = pre1 + df['match_no']

# Adding prefix 'm_' to 'match_no' column in 'df1'
df1['match_no'] = pre2 + df1['match_no']
```

```python
df1
```

```python
df.loc[df['home_or_away']=='H','Team']=df['home_team']
df.loc[df['home_or_away']=='A','Team']=df['away_team']
df.loc[df['home_or_away']=='N','Team']=df['home_team']
df1.loc[df1['home_or_away']=='H','Team']=df1['home_team']
df1.loc[df1['home_or_away']=='A','Team']=df1['away_team']
df1.loc[df1['home_or_away']=='N','Team']=df1['home_team']
```

```python
both=pd.concat([df,df1],axis=0)
```

```python
both.to_csv('M vs R.csv',index=False,encoding='utf-8-sig')
df.to_csv('Ronaldo.csv',index=False,encoding='utf-8-sig')
df1.to_csv('Messi.csv',index=False,encoding='utf-8-sig')
```
